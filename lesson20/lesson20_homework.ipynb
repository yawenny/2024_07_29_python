```python
import requests
from requests import ConnectionError,HTTPError,Timeout
url = 'https://data.ntpc.gov.tw/api/datasets/010E5B15-3823-4B20-B401-B1CF000550C5/json?page=0&size=2000'
try:
    response = requests.get(url)    
    response.raise_for_status()    
except ConnectionError:
    print('找不到伺服器')
except HTTPError:
    print('網頁找不到')
except Timeout:
    print('超過時間沒有回應')
else:
    print('沒有發生問題')

import pandas as pd
df1 = pd.DataFrame(response.json(),
                   columns=['sna','tot','sbi','sarea','ar','bemp'])

df2 = df1.rename(columns={'sna':'站名', 'sarea':'區域', 'ar':'地址', 'tot':'數量','sbi':'可借', 'bemp':'可還'})

#object轉成int
df2['數量'] = df2['數量'].map(lambda value:int(value))
df2['可還'] = df2['可還'].map(lambda value:int(value))
df2['可借'] = df2['可借'].map(lambda value:int(value))

#使用'區域'欄位進行groupby及統計
groupby = df2.groupby(by='區域')
df3 = groupby[['數量','可還','可借']].sum()
df3
```
