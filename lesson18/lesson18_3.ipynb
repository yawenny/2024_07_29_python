```python
#index和columns的名字
import numpy as np
import pandas as pd

scores = np.random.randint(50,101,size=(50,5))
scores_df =pd.DataFrame(scores,
             columns=['國文','英文','數學','地理','歷史'],
             index=range(1,51))
type(scores_df.index)
scores_df.index.name='學號'
scores_df.columns.name='科目'
scores_df
```

```python
#將每一欄series計算統計後,再合併為dataframe
def call(s):
    min = s.min()
    max = s.max()
    median = s.median()
    return pd.Series([min,max,median],index=['最低分','最高分','中間值'])

scores_df.apply(call)
```

```python
#列series,及格及不及格的數量
def call1(s):
    less60 = s[s<60]
    return less60.count()

less60_s = scores_df.apply(call1,axis=1)
scores_df['不及格的科目']=less60_s
scores_df
```
