```python
import pandas as pd
file_path ='每日各站進出站人數/每日各站進出站人數2020.csv'
df1 = pd.read_csv(file_path)
df1
```

```python
#欄位改名
df2 = df1.rename(columns = {"trnOpDate":"乘車日",
          "staCode" :"車站代碼",
           "gateInComingCnt":"進站人數",
            "gateOutGoingCnt":"出站人數" })
df2
```

```python
#車站代碼及車站名dataframe
import json
from pprint import pprint

with open('車站基本資料集.json',encoding= 'utf-8') as file:
    json_date =json.load(file)#json變成Python資料結構list of dict
    
stations_info =pd.DataFrame(json_date,columns=['stationCode','stationName'])#變成dataframe
#從字串object轉成整數,且覆蓋
stations_info['stationCode'] = stations_info['stationCode'].map(lambda value:int(value))
#變int64
stations_info.info()
```

```python
#車站代碼及車站名dataframe
import json
from pprint import pprint

with open('車站基本資料集.json',encoding= 'utf-8') as file:
    json_date =json.load(file)#json變成Python資料結構list of dict
    
stations_info =pd.DataFrame(json_date,columns=['stationCode','stationName'])#變成dataframe

#將欄位改type,object變int
stations_info = stations_info.astype({'stationCode':'int32'})
stations_info
```

```python
import json
from pprint import pprint

with open('車站基本資料集.json',encoding='utf-8') as file:
   json_data = json.load(file)

stations_info = pd.DataFrame(json_data,columns=['stationCode','stationName'])
stations_info['stationCode'] = stations_info['stationCode'].map(lambda value:int(value))
stations_info.info()
```

```python
#join二個dataframe,用merge(),pandas的function
df3 = pd.merge(df2,stations_info,left_on="車站代碼",right_on='stationCode')
df4 = df3.rename(columns={'stationName':'車站名稱'})
df4
```

```python
#df2.drop(columns=['車站代碼','stationCode'])
df5=df4.reindex(columns=['乘車日','車站名稱','進站人數','出站人數'])
df5
```

```python
#先轉日期為字串,再變為datetime格式
df5['乘車日'] = pd.to_datetime(df5['乘車日'].astype(str))
df5.info()
```
