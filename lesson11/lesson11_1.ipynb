```python
#下載網頁資料
import requests
from requests.exceptions import HTTPError,ConnectionError,Timeout
try:
    url = 'https://opendata.cwa.gov.tw/fileapi/v1/opendataapi/O-A0091-001?Authorization=rdec-key-123-45678-011121314&format=JSON'
    response = requests.get(url)
    response.raise_for_status()
    #print(type(response))
    content = response.text
    #print(content)
except ConnectionError:
    print("連線伺服器異常")
except Timeout:
    print("伺服器太忙,沒有回應")
except HTTPError:
    print("伺服器回應連線錯誤")
except Exception:
    print("不知名的錯誤")
```

```python
import io
import json
from pprint import pprint
#純文字變成file物件;不用存檔方式
file = io.StringIO(content)
#print(type(file))
#從file物件變成python的dictionary
contents = json.load(file)
```

```python
#type(contents)
#cwaopendata是dict,sent是key,其value是字串
date:str = contents['cwaopendata']['sent']
pprint(date)
```

```python
#cwaopendata及dataset是dict,station是list,下面有很多dict及key及value,到了station這一層
stations:list = contents['cwaopendata']['dataset']['Station']
pprint(stations)
```

```python
#資料太複雜,用class來做類別,有2個屬性name,weather
class Station():
    def __init__(self,name:str,weather:float):
        self.name = name
        self.weather = weather

#station層下有StationName及['WeatherElement']['SolarRadiation']
stations1:list=[]
for station in stations:
    name = (station['StationName'])
    weather:str = station['WeatherElement']['SolarRadiation']
    station:Station = Station(name,float(weather))
    stations1.append(station)

pprint(stations1)
```
