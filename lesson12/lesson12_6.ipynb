```python
import requests
from requests import Response
from pydantic import BaseModel,RootModel,Field,field_validator,field_serializer
from datetime import datetime
from pprint import pprint
url ='https://data.ntpc.gov.tw/api/datasets/c51d5111-c300-44c9-b4f1-4b28b9929ca2/json?size=1000' 
try:
    responses:Response = requests.request('GET',url)
    responses.raise_for_status()
except Exception:
    print("下載錯誤")
else:
    content = responses.text

class Factory(BaseModel):
    名稱:str = Field(alias='organizer')
    註冊號碼:str = Field(alias='no')
    地址:str = Field(alias='address')
    統編:str = Field(alias='tax_id_number')
    經度:float|None = Field(alias='wgs84ax')
    緯度:float|None = Field(alias='wgs84ay')
    日期:datetime= Field(alias='date')

    @field_validator('經度','緯度',mode='before')
    @classmethod
    def slice_float(cls,number:float)->float:
        if number==None:
            return None
        else:
            return format(float(number),'.5f')     
    
    @field_serializer('日期')
    def date_serial(self,date:datetime)->str:
        return date.strftime(f"中華民國{date.year-1911}年%m月%d日")

class Factory_data(RootModel):
    root:list[Factory]


factories:Factory_data = Factory_data.model_validate_json(content)
factories_list = factories.model_dump()
pprint(factories_list)
```
